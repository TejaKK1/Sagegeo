Bivariate Normal Distribution
X,Y ~ BVN(u1, u2, s1, s2, r)

```python
var('x', 'y')
u1=0
u2=0
s1=1
s2=1
r=0
f(x, y) = e^(-1/(2*(1-r^2))*(((x-u1)/s1)^2+((y-u2)/s2)^2-2*r*((x-u1)/s1)*((y-u2)/s2)))/(2*pi*s1*s2*sqrt(1-r^2))
plot3d(f,(-3,3),(-3,3),(0,0))
```
