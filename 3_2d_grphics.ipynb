```python
var('x,y')
plt = []
# Define a color dictionary
color = {
    0: 'red',
    1: 'blue',
    2: 'green',
    3: 'purple',
    4: 'orange',
    5: 'cyan',
    6: 'magenta',
    7: 'yellow',
    8: 'black',
    9: 'brown',
    10: 'pink'
}
for i in range(1, 6):
    n = i
    f(x,y) = x^n + y^n - 1
    color_index = (10 + i^2) % len(color)  # Use modulo to wrap around the colors
    p = implicit_plot(f, (x, -2, 2), (y, -2, 2), color=color[color_index])
    plt.append(p)
    
# Display all plots
show(sum(plt))
```

```python
g(x) = x-3
y = var('y')
eq = x == y^2 + 2*y -3
sol = solve(eq,y)
h1(x) = sol[0].rhs()
h2(x) = sol[1].rhs()

G = Graphics()
G += plot(g,-5,5, fill=g(5),fillcolor='gold')
G += plot(h1,-4,5)
G += plot(h2,-4,5,fill=h1,fillcolor='blue',fillalpha=0.2)
G.show()
```
